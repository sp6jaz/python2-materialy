## Problem NumPy — brak kontekstu

Zanim poznamy Pandas, zobaczmy **dlaczego** go potrzebujemy.
NumPy daje nam szybkość — ale nie daje kontekstu.

````python
# NumPy — szybki, ale ślepy
sprzedaz_np = np.array([340, 120, 560, 90, 420])
print(f"Sprzedaż: {sprzedaz_np}")
print()
print("Pytania bez odpowiedzi:")
print("  340 czego? Sztuk? Złotówek?")
print("  Który to produkt?")
print("  Który to miesiąc?")
print("  NumPy nie wie. Ma tylko liczby.")
````

````python
# Pandas — te same dane, ale Z KONTEKSTEM
sprzedaz_pd = pd.Series(
    [340, 120, 560, 90, 420],
    index=['Laptop', 'Tablet', 'Smartfon', 'Akces.', 'Monitor']
)
print(sprzedaz_pd)
print()
print("Teraz widać: Laptop — 340 szt., Smartfon — 560 szt.")
print("Dane z kontekstem. To jest siła Pandas.")
````

### Pandas vs NumPy — podsumowanie różnic

| Cecha | NumPy | Pandas |
|-------|-------|--------|
| Struktura | `ndarray` (tablica) | `Series` (kolumna) / `DataFrame` (tabela) |
| Indeks | Tylko numeryczny (0, 1, 2...) | Dowolne etykiety (nazwy, daty...) |
| Typy w kolumnie | Jeden typ w całej tablicy | Każda kolumna może mieć inny typ |
| Wczytywanie CSV | Brak (tylko `np.loadtxt`) | `pd.read_csv()` — jedna linia |
| Brakujące dane | Brak wbudowanego wsparcia | `NaN`, `isna()`, `fillna()`, `dropna()` |
| Szybkość | Szybszy (mniej narzutu) | Wolniejszy, ale wygodniejszy |

**Reguła:** NumPy do obliczeń numerycznych, Pandas do pracy z danymi tabelarycznymi.

## 1. Series — tablica z etykietami

**Series** to podstawowa struktura Pandas. To jednowymiarowa tablica (jak w NumPy),
ale z **indeksem** — każdy element ma swoją etykietę.

Wyobraź sobie kolumnę w arkuszu Excel:
- Wartości: 340, 120, 560...
- Nagłówki wierszy: Laptop, Tablet, Smartfon...

To właśnie Series.

| Sposób tworzenia | Przykład | Kiedy używać |
|-------------------|---------|-------------|
| Z listy | `pd.Series([1, 2, 3])` | Szybkie dane testowe |
| Z listy + index | `pd.Series([1, 2, 3], index=['a','b','c'])` | Dane z etykietami |
| Ze słownika | `pd.Series({'a': 1, 'b': 2})` | Najnaturalniejszy sposób |
| Z tablicy NumPy | `pd.Series(np_array)` | Konwersja z NumPy |

**Dokumentacja:** [pd.Series](https://pandas.pydata.org/docs/reference/api/pandas.Series.html)

````python
# Tworzenie Series z listy — Pandas automatycznie dodaje indeks 0, 1, 2...
s1 = pd.Series([340, 120, 560, 90, 420])
print(s1)
print()
print(f"Typ: {type(s1)}")
print(f"Indeks: {list(s1.index)}")
print(f"Wartości: {s1.values}")   # pod spodem to tablica NumPy!
print(f"Typ wartości: {type(s1.values)}")
````

````python
# Series z własnymi etykietami (index)
sprzedaz = pd.Series(
    [340, 120, 560, 90, 420],
    index=['Laptop', 'Tablet', 'Smartfon', 'Akces.', 'Monitor']
)
print(sprzedaz)
print()
print("Teraz każda wartość ma swoją nazwę — nie trzeba pamiętać,")
print("że element [2] to Smartfon.")
````

### Dostęp do elementów Series

Dostęp po **etykiecie**: `s['Laptop']` — znam nazwę.
Dostęp po **pozycji**: `s.iloc[0]` — znam numer.

Na **W06** poznamy `loc` i `iloc` dokładnie — dziś wystarczy wiedzieć, że istnieją.

````python
# Dostęp do elementów Series
print(f"Po etykiecie: sprzedaz['Smartfon'] = {sprzedaz['Smartfon']}")
print(f"Po pozycji:   sprzedaz.iloc[0] = {sprzedaz.iloc[0]}")
print()
print("Pierwsze 3 elementy (slice):")
print(sprzedaz[:3])
````

### Operacje na Series

Series wspiera **te same operacje wektorowe** co NumPy:
mnożenie, dodawanie, filtrowanie boolean, agregacje.
Ale z bonusem — wyniki zachowują etykiety!

````python
# Operacje wektorowe — działają identycznie jak w NumPy!
print("=== Mnożenie przez skalar (ceny z VAT) ===")
sprzedaz_vat = sprzedaz * 1.23
print(sprzedaz_vat.round(0))
print()

print("=== Dodawanie skalara (premia +50 szt.) ===")
print(sprzedaz + 50)
````

````python
# Filtrowanie boolean — jak w NumPy, ale z etykietami
print("Produkty ze sprzedażą > 200:")
maska = sprzedaz > 200
print(f"Maska boolean:\n{maska}")
print()
popularne = sprzedaz[maska]
print(f"Wynik filtrowania:")
print(popularne)
print(f"\nIle popularnych: {len(popularne)} z {len(sprzedaz)}")
````

````python
# Agregacje — jak NumPy, ale z bonusem
print(f"Średnia sprzedaż:   {sprzedaz.mean():.0f} szt.")
print(f"Suma:              {sprzedaz.sum()} szt.")
print(f"Min:               {sprzedaz.min()} szt. ({sprzedaz.idxmin()})")
print(f"Max:               {sprzedaz.max()} szt. ({sprzedaz.idxmax()})")
print(f"Mediana:           {sprzedaz.median():.0f} szt.")
print()
print("Bonus Pandas: idxmax() i idxmin() zwracają ETYKIETĘ, nie numer!")
print(f"  NumPy argmax():  {sprzedaz.values.argmax()}  — numer pozycji")
print(f"  Pandas idxmax(): {sprzedaz.idxmax()}  — nazwa produktu")
````

### Tworzenie Series ze słownika i z NumPy

Słownik Pythona to **najnaturalniejszy** sposób tworzenia Series:
klucze stają się indeksem, wartości stają się danymi.

````python
# Series ze słownika — najnaturalniejszy sposób
populacja = pd.Series({
    'Warszawa':  1_794_000,
    'Kraków':    800_000,
    'Wrocław':   643_000,
    'Poznań':    534_000,
    'Opole':     128_000
})
print(populacja)
print()
print(f"Opole: {populacja['Opole']:,} mieszkańców")
print("Klucze dict → indeks, wartości dict → dane")
````

````python
# Series z tablicy NumPy — konwersja
dane_np = np.array([10.5, 20.3, 15.7, 18.2])
dane_pd = pd.Series(dane_np, index=['Q1', 'Q2', 'Q3', 'Q4'])
print(dane_pd)
print()
print("NumPy → Pandas: dodajemy etykiety do istniejących danych")
print(f"Typ wartości: {dane_pd.dtype}  — to nadal float64 z NumPy")
````

### Sortowanie i unikalne wartości

Series ma wbudowane metody do sortowania i analizy wartości.

````python
# Sortowanie
print("=== sort_values() — sortowanie po wartościach ===")
print(sprzedaz.sort_values())          # rosnąco
print()
print("=== sort_values(ascending=False) — malejąco ===")
print(sprzedaz.sort_values(ascending=False))
print()
print("=== sort_index() — sortowanie po indeksie (alfabetycznie) ===")
print(sprzedaz.sort_index())
````

````python
# Unikalne wartości — analiza kategorii produktów
zamowienia = pd.Series(['Laptop', 'Tablet', 'Laptop', 'Smartfon',
                        'Laptop', 'Tablet', 'Smartfon', 'Laptop'])
print(f"Zamówienia: {list(zamowienia)}")
print(f"Unikalne produkty: {list(zamowienia.unique())}")
print(f"Ile unikalnych: {zamowienia.nunique()}")
print()
print("Rozkład zamówień:")
print(zamowienia.value_counts())
````

### Bonus: zaawansowane cechy Series

Poniższe tematy nie są wymagane dziś — ale warto wiedzieć, że istnieją.

### Bonus: atrybut `name`

Series może mieć **nazwę** — to będzie nazwa kolumny,
gdy Series trafi do DataFrame.

````python
# name — nazwa Series
sprzedaz.name = 'sprzedaz_szt'
print(sprzedaz)
print()
print(f"Nazwa: {sprzedaz.name}")
print(f"Indeks nazwa: {sprzedaz.index.name}")
print()
print("Nazwa Series → nazwa kolumny w DataFrame.")
````

### Bonus: arytmetyka między dwoma Series

Gdy wykonujesz operację na dwóch Series, Pandas automatycznie
**dopasowuje elementy po indeksie** (etykiecie), nie po pozycji!
Jeśli jakiegoś klucza brakuje — wynik to `NaN` (brak danych).

````python
# Arytmetyka między Series — dopasowanie po indeksie
przychod_q1 = pd.Series({'Laptop': 340, 'Tablet': 120, 'Smartfon': 560})
przychod_q2 = pd.Series({'Smartfon': 600, 'Laptop': 380, 'Monitor': 200})

print("Q1:")
print(przychod_q1)
print("\nQ2:")
print(przychod_q2)
print("\nQ1 + Q2 (dopasowanie po indeksie):")
print(przychod_q1 + przychod_q2)
print()
print("Monitor: był tylko w Q2 → NaN (brak w Q1)")
print("Tablet: był tylko w Q1 → NaN (brak w Q2)")
print("Pandas NIE zgaduje — jeśli nie ma danych, mówi NaN.")
````

## 2. DataFrame — tabela z wierszami i kolumnami

**DataFrame** to główna struktura danych w Pandas. To tabela — jak arkusz w Excelu:
- **Wiersze** — obserwacje (klienci, produkty, transakcje)
- **Kolumny** — cechy/zmienne (nazwa, cena, ilość)
- Każda kolumna to **Series**

```
          produkt    cena  sprzedaz  kategoria
    0     Laptop     3500      340  Komputery
    1     Tablet     1800      120  Mobilne
    2     Smartfon   2500      560  Mobilne
    ↑          ↑        ↑        ↑         ↑
 indeks   kolumna  kolumna  kolumna   kolumna
```

**Dokumentacja:** [pd.DataFrame](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html)

### Tworzenie DataFrame

Najczęstsze sposoby:

| Sposób | Opis | Kiedy |
|--------|------|-------|
| `pd.DataFrame({'kol': [dane]})` | Ze słownika list | Tworzenie ręcznie |
| `pd.DataFrame([{}, {}, ...])` | Z listy słowników | Dane z API/JSON |
| `pd.read_csv('plik.csv')` | Z pliku CSV | 90% przypadków |
| `pd.read_excel('plik.xlsx')` | Z Excela | Dane firmowe |

````python
# Tworzenie DataFrame ze słownika list
# Klucze = nazwy kolumn, wartości = dane w kolumnach
dane = pd.DataFrame({
    'produkt':   ['Laptop', 'Tablet', 'Smartfon', 'Akces.', 'Monitor'],
    'cena':      [3500, 1800, 2500, 150, 2200],
    'sprzedaz':  [340, 120, 560, 90, 420],
    'kategoria': ['Komputery', 'Mobilne', 'Mobilne', 'Akcesoria', 'Komputery']
})
print(dane)
````

````python
# DataFrame z listy słowników — np. dane z API
# Każdy słownik to jeden wiersz
zamowienia = pd.DataFrame([
    {'klient': 'Anna',  'produkt': 'Laptop',   'kwota': 3500},
    {'klient': 'Jan',   'produkt': 'Tablet',   'kwota': 1800},
    {'klient': 'Kasia', 'produkt': 'Smartfon', 'kwota': 2500},
])
print(zamowienia)
print()
print("Ten format często spotkacie przy danych z API (JSON).")
````

````python
# Atrybuty DataFrame — „metryczka” danych
print(f"Kształt (wiersze, kolumny): {dane.shape}")    # (5, 4)
print(f"Kolumny: {list(dane.columns)}")                 # nazwy kolumn
print(f"Indeks: {dane.index}")                          # RangeIndex(0, 5)
print(f"Rozmiar: {dane.size}")                          # 20 (5×4)
print(f"Wymiary: {dane.ndim}")                          # 2
````

### Typy danych (`dtypes`)

Każda kolumna w DataFrame ma swój **typ danych**. To fundamentalna cecha Pandas
i trzeba ją rozumieć, bo **typ danych decyduje, jakie operacje są możliwe**.

| Typ Pandas | Opis | Przykład | Możliwe operacje |
|------------|------|---------|------------------|
| `int64` | Liczby całkowite | 1, 42, -5 | Arytmetyka, mean, sum |
| `float64` | Liczby zmiennoprzecinkowe | 3.14, -0.5 | Arytmetyka, mean, sum |
| `object` | Tekst (string) | 'Laptop', 'Tak' | value_counts, unique |
| `bool` | Wartości logiczne | True, False | Filtrowanie |
| `datetime64` | Data i czas | 2026-03-21 | Analiza czasowa |
| `category` | Kategorie (oszczędza pamięć) | 'M', 'F' | value_counts, groupby |

**Reguła:** Nie możesz policzyć średniej z tekstu! Dlatego **zawsze sprawdzaj typy** na początku.

````python
# Typy danych w naszym DataFrame
print(dane.dtypes)
print()
print("produkt → object (tekst)  — nie policzysz średniej")
print("cena → int64 (liczba)     — można liczyć mean, sum, max")
print("sprzedaz → int64          — to samo")
print("kategoria → object (tekst) — value_counts, unique")
````

### Wczytywanie danych z pliku CSV

W praktyce nie wpisujecie danych ręcznie. **90% pracy** to wczytanie gotowego pliku.

`pd.read_csv()` wczytuje plik CSV (Comma-Separated Values) do DataFrame.
Może to być plik lokalny lub URL.

**Najważniejsze parametry:**

| Parametr | Opis | Przykład |

| `nrows` | Ile wierszy wczytać | `nrows=100` — do testów |

**Dokumentacja:** [pd.read_csv](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)

````python
# Wczytanie prawdziwego datasetu z URL
# Dataset 'tips' — rachunki i napiwki w amerykańskiej restauracji
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
tips = pd.read_csv(url)

print(f"Wczytano: {tips.shape[0]} wierszy, {tips.shape[1]} kolumn")
print(f"Kolumny: {list(tips.columns)}")
print()
print("Jedna linia kodu — i mamy 244 rachunki z restauracji!")
````

## 3. Eksploracja danych (EDA)

**EDA — Exploratory Data Analysis** (eksploracyjna analiza danych) — to pierwsza rzecz,
którą robisz z każdym nowym zestawem danych. Zanim odpowiesz na jakiekolwiek pytanie
biznesowe — musisz znać swoje dane.

**5 metod, które odpalasz na KAŻDYM nowym datasecie:**

| # | Metoda | Co robi | Analogia |
|---|--------|---------|----------|
| 1 | `df.head()` | Pierwsze wiersze | Otwieram plik i patrzę |
| 2 | `df.info()` | RTG danych — typy, rozmiar, braki | Sprawdzam „metryczkę” |
| 3 | `df.describe()` | Statystyki liczbowe | Obliczam podstawowe liczby |
| 4 | `df['kol'].value_counts()` | Rozkład kategorii | Liczę ile czego |
| 5 | `df.isna().sum()` | Ile braków | Sprawdzam puste komórki |

Zapamiętaj te 5 metod — to twoja **checklista na start** każdej analizy.

### `info()` — RTG danych

Najważniejsza metoda eksploracyjna. Jeden rzut oka — widzisz wszystko:
rozmiar, typy, braki.

````python
# info() — RTG danych
tips.info()
````

**Jak czytać `info()`:**

- `244 entries` — 244 wiersze (rachunki)
- `7 columns` — 7 kolumn (cech)
- Dla każdej kolumny: nazwa, ile wartości non-null, typ danych
- `244 non-null` wszędzie — brak braków (w realnych danych tak pięknie nie będzie!)
- Na dole: ile pamięci zajmują dane

**Wniosek:** Dane kompletne, 3 kolumny liczbowe, 4 tekstowe.

### `describe()` — statystyki opisowe

Podsumowanie statystyczne **kolumn liczbowych**. Jedno wywołanie — i znasz
rozkład danych.

````python
# describe() — statystyki opisowe kolumn liczbowych
print(tips.describe())
````

**Jak czytać `describe()`:**

| Wiersz | Znaczenie | Przykład (total_bill) |
|--------|-----------|------------------------|
| `count` | Ile wartości (bez NaN) | 244 |
| `mean` | Średnia | ~19.79 USD |
| `std` | Odchylenie standardowe | ~8.90 USD |
| `min` | Minimum | 3.07 USD |
| `25%` | Kwartyl 1 (Q1) — 25% danych poniżej | ~13.35 USD |
| `50%` | Mediana (Q2) — środek rozkładu | ~17.80 USD |
| `75%` | Kwartyl 3 (Q3) — 75% danych poniżej | ~24.13 USD |
| `max` | Maksimum | 50.81 USD |

**Co widzimy?** Średni rachunek ~20 USD, napiwek ~3 USD, grupy 1–6 osób.
Mediana (17.80) < średnia (19.79) — rozkład jest **prawoskośny** (kilka dużych rachunków podnosi średnią).

````python
# describe(include='all') — statystyki WSZYSTKICH kolumn (też tekstowych)
print(tips.describe(include='all'))
````

**Kolumny tekstowe** (`object`) mają inne statystyki:

| Wiersz | Znaczenie | Przykład (day) |
|--------|-----------|----------------|
| `unique` | Ile unikalnych wartości | 4 (Thur, Fri, Sat, Sun) |
| `top` | Najczęstsza wartość | Sat |
| `freq` | Ile razy wystąpiła | 87 |

Dzięki temu od razu widzisz: sobota to najczęstszy dzień, obiad to dominująca pora.

### `value_counts()` — rozkład wartości

`value_counts()` to odpowiednik funkcji **LICZ.JEŻELI** w Excelu —
ile razy każda wartość wystąpiła w kolumnie.

````python
# value_counts() — rozkład wartości w kolumnie
print("=== Dzień tygodnia ===")
print(tips['day'].value_counts())

print("\n=== Pora dnia ===")
print(tips['time'].value_counts())

print("\n=== Palacz? ===")
print(tips['smoker'].value_counts())
````

````python
# value_counts(normalize=True) — proporcje zamiast liczebności
print("Proporcje (dzień tygodnia):")
print(tips['day'].value_counts(normalize=True).round(2))
print()
print("Sobota = 36%, niedziela = 31%, czwartek = 25%, piątek = 8%")
print("Przydatne w raportach — procenty są czytelniejsze niż surowe liczby.")
````

> **Ciekawostka z danych:** Piątek to tylko 8% rachunków — dlaczego?
> Restauracja prawdopodobnie była zamknięta w poniedziałek–środę
> (brak tych dni w danych). Piątek to dzień otwarcia — mniej gości niż w weekend.
> To właśnie robi EDA: **dane opowiadają historię**, trzeba tylko umieć czytać.

````python
# Kto płaci rachunek? Rozkład po płci
print("Kto płaci rachunek?")
print(tips['sex'].value_counts())
print()
print(tips['sex'].value_counts(normalize=True).round(2))
print()
print("Mężczyźni płacą 64% rachunków. Dane z lat 90. w USA.")
````

### `isna()` — brakujące wartości

W realnych danych **zawsze** są braki — puste komórki, brak odpowiedzi w ankiecie,
błędy w imporcie. Sprawdzenie braków to **obowiązkowy** krok EDA.

Pandas oznacza brakujące wartości jako `NaN` (Not a Number — dosł. „nie liczba”).

````python
# isna() — sprawdzenie brakujących wartości
# ZAWSZE rób to jako jeden z pierwszych kroków!
print("Brakujące wartości per kolumna:")
print(tips.isna().sum())
print(f"\nŁącznie braków: {tips.isna().sum().sum()}")
print()
print("Jak to działa:")
print("  isna()  — tworzy tablicę True (brak) / False (jest)")
print("  .sum()  — liczy ile True w każdej kolumnie")
print("  Tu 0 wszędzie — dane kompletne. Na W07 nauczycie się radzić z brakami.")
````

### `nlargest()` i `nsmallest()` — top N

Szybki sposób na znalezienie rekordów z największymi/najmniejszymi wartościami.

````python
# nlargest() — top 5 największych rachunków
print("Top 5 największych rachunków:")
print(tips.nlargest(5, 'total_bill')[['total_bill', 'tip', 'day', 'size']])
print()

# nsmallest() — 5 najmniejszych rachunków
print("5 najmniejszych rachunków:")
print(tips.nsmallest(5, 'total_bill')[['total_bill', 'tip', 'day', 'size']])
````

````python
# Ile unikalnych wartości w każdej kolumnie?
print("Unikalne wartości per kolumna:")
print(tips.nunique())
print()
print("total_bill: 229 unikalnych z 244 — prawie każdy rachunek inny")
print("day: 4 — czwartek, piątek, sobota, niedziela")
print("time: 2 — lunch, dinner")
print("size: 6 — grupy od 1 do 6 osób")
````

## 4. Selekcja kolumn i obliczenia

Jak wybrać konkretną kolumnę z DataFrame?

| Składnia | Wynik | Przykład |
|---------|-------|----------|
| `df['nazwa']` | **Series** (jedna kolumna) | `tips['total_bill']` |
| `df[['a', 'b']]` | **DataFrame** (wiele kolumn) | `tips[['total_bill', 'tip']]` |

**UWAGA na nawiasy:** `df['a']` → Series. `df[['a', 'b']]` → DataFrame.
Podwójne nawiasy `[['a', 'b']]` — wewnątrz jest **lista** nazw kolumn.

````python
# Jedna kolumna → Series
rachunki = tips['total_bill']
print(f"Typ: {type(rachunki).__name__}")
print(rachunki.head())
print()
print("Każda kolumna DataFrame to Series — ze swoim indeksem i typem danych.")
````

### Jedna kolumna vs wiele kolumn — dlaczego to ważne?

To rozróżnienie jest **kluczowe**, bo metody działają inaczej:
- **Series** (jedna kolumna): `.mean()`, `.max()`, `.value_counts()` — działają wprost
- **DataFrame** (wiele kolumn): `.mean()` liczy średnią **każdej kolumny** osobno

Typowy błąd początkującego: `tips['total_bill', 'tip']` — **brak podwójnych nawiasów**.
Poprawnie: `tips[['total_bill', 'tip']]`

````python
# Wiele kolumn → DataFrame
pieniadze = tips[['total_bill', 'tip']]
print(f"Typ: {type(pieniadze).__name__}")
print(pieniadze.head())
print()
print("Podwójne nawiasy: df[['a', 'b']] — wewnątrz jest lista ['a', 'b']")
````

### Tworzenie nowej kolumny

Nową kolumnę tworzymy jak zwykłe przypisanie — Pandas robi obliczenia
element po elemencie (jak NumPy).

````python
# Tworzenie nowej kolumny — obliczenia na istniejących
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)

print("Nowa kolumna: procent napiwku")
print(tips[['total_bill', 'tip', 'tip_pct']].head())
print()
print(f"Średni napiwek: {tips['tip_pct'].mean():.1f}%")
print(f"Max napiwek:   {tips['tip_pct'].max():.1f}%")
print(f"Min napiwek:   {tips['tip_pct'].min():.1f}%")
````

````python
# Kto dał najwyższy napiwek procentowo?
idx_max = tips['tip_pct'].idxmax()
print(f"Rekordowy napiwek — wiersz {idx_max}:")
print(tips.loc[idx_max])
print()
print(f"Rachunek: {tips.loc[idx_max, 'total_bill']} USD")
print(f"Napiwek:  {tips.loc[idx_max, 'tip']} USD ({tips.loc[idx_max, 'tip_pct']}%)")
````

**Co tu się stało krok po kroku:**
1. `tips['tip_pct'].idxmax()` — znalazł **numer wiersza** z najwyższym napiwkiem %
2. `tips.loc[idx]` — wyciągnął **cały wiersz** po tym numerze
3. `tips.loc[idx, 'total_bill']` — wyciągnął **konkretną wartość** z tego wiersza

Ten wzorzec „znajdź indeks → wyciągnij dane” będziecie stosować bardzo często.

### Pod spodem to NumPy!

Pandas jest zbudowany **na NumPy**. Każda kolumna (Series) pod spodem
to tablica `numpy.ndarray`. Możesz ją wyciągnąć przez `.values`.

````python
# Pod spodem to NumPy!
wartosci = tips['total_bill'].values
print(f"Typ .values: {type(wartosci)}")   # numpy.ndarray
print(f"Dtype: {wartosci.dtype}")          # float64
print()

# Funkcje NumPy działają na kolumnach Pandas
print(f"np.median:     {np.median(tips['total_bill']):.2f}")
print(f"Pandas median: {tips['total_bill'].median():.2f}")
print()
print("Oba dają ten sam wynik. Używajcie metod Pandas — wygodniejsze.")
````

````python
# Sortowanie DataFrame
print("=== Top 5 rachunków (malejąco) ===")
print(tips.sort_values('total_bill', ascending=False).head())
print()
print("sort_values('kolumna') sortuje cały DataFrame po wybranej kolumnie.")
print("Więcej o sortowaniu i filtrowaniu na W06.")
````

| `df.sample(n)` | Losowe n wierszy |
| `df.shape` | (wiersze, kolumny) |
| `df.columns` | Nazwy kolumn |
| `df.dtypes` | Typy danych |
| `df.info()` | RTG danych — rozmiar, typy, braki |
| `df.describe()` | Statystyki kolumn liczbowych |
| `df.describe(include='all')` | Statystyki wszystkich kolumn |
| `df['kol'].value_counts()` | Zliczanie wartości |
| `df['kol'].value_counts(normalize=True)` | Proporcje (%) |
| `df.isna().sum()` | Liczba braków per kolumna |
| `df.nunique()` | Ile unikalnych wartości |
| `df.nlargest(n, 'kol')` | Top n wierszy |

### Selekcja kolumn

|---------|-------|
